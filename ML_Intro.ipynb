**Question 1: Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).**

**Answer:**

Artificial Intelligence (AI), Machine Learning (ML), Deep Learning (DL), and Data Science (DS) are closely related fields, but each has its own scope, techniques, and applications.

**1. Artificial Intelligence (AI):**

* **Definition:** AI is the broadest field among all. It refers to the simulation of human intelligence in machines that are programmed to think, reason, and act like humans.
* **Scope:** AI includes everything that makes a system "smart," such as problem-solving, reasoning, natural language understanding, planning, and robotics.
* **Techniques Used:** Rule-based systems, expert systems, ML algorithms, neural networks, search algorithms.
* **Applications:**

  * Voice assistants like Siri, Alexa, Google Assistant.
  * Self-driving cars (Tesla Autopilot).
  * Game playing (Chess, Go by AlphaGo).
* **Key Point:** AI is the umbrella term under which ML and DL exist.

**2. Machine Learning (ML):**

* **Definition:** ML is a subset of AI that focuses on giving machines the ability to learn from data and improve over time without being explicitly programmed.
* **Scope:** Instead of hardcoding rules, ML systems use algorithms to detect patterns and make predictions or decisions.
* **Techniques Used:** Regression, Classification, Clustering, Decision Trees, Random Forests, Support Vector Machines.
* **Applications:**

  * Email spam filtering.
  * Product recommendations on Amazon/Netflix.
  * Fraud detection in banking.
* **Key Point:** ML narrows AI's focus to data-driven learning.

**3. Deep Learning (DL):**

* **Definition:** DL is a specialized subset of ML that uses Artificial Neural Networks (ANNs) with many layers (hence “deep”). It mimics how the human brain processes information.
* **Scope:** It works best with large datasets and high computational power. Unlike traditional ML, DL automatically extracts features from raw data (no need for manual feature engineering).
* **Techniques Used:** Convolutional Neural Networks (CNNs), Recurrent Neural Networks (RNNs), Generative Adversarial Networks (GANs).
* **Applications:**

  * Image recognition (Facebook face tagging, medical imaging).
  * Speech recognition (Google Translate, voice assistants).
  * Autonomous vehicles (object detection and navigation).
* **Key Point:** DL is ML on steroids—designed for big, complex data like images, video, and natural language.

**4. Data Science (DS):**

* **Definition:** Data Science is an interdisciplinary field that combines programming, statistics, mathematics, and domain knowledge to extract meaningful insights from data.
* **Scope:** It is broader than AI/ML/DL because it includes the entire process of collecting, cleaning, analyzing, and visualizing data, along with applying ML models when needed.
* **Techniques Used:** Data cleaning, Exploratory Data Analysis (EDA), visualization (Matplotlib, Power BI), statistical modeling, and ML.
* **Applications:**

  * Business forecasting (sales prediction).
  * Social media analytics (trends, sentiment analysis).
  * Healthcare analytics (disease prediction, patient monitoring).
* **Key Point:** DS is about understanding data, not just building intelligent systems.

**5. Comparison Table:**

| Aspect           | Artificial Intelligence (AI)                 | Machine Learning (ML)                             | Deep Learning (DL)                 | Data Science (DS)                                            |
| ---------------- | -------------------------------------------- | ------------------------------------------------- | ---------------------------------- | ------------------------------------------------------------ |
| **Definition**   | Broad concept of making machines intelligent | Subset of AI, teaches machines to learn from data | Subset of ML using neural networks | Interdisciplinary field for analyzing and understanding data |
| **Scope**        | Covers all intelligent behavior              | Focuses on algorithms                             | Focuses on neural networks         | Covers entire data pipeline                                  |
| **Data Need**    | Varies                                       | Moderate                                          | Requires huge data                 | Structured + unstructured data                               |
| **Techniques**   | Logic, rules, ML, DL                         | Regression, SVM, Trees                            | CNNs, RNNs, GANs                   | Statistics, ML, visualization                                |
| **Applications** | Robots, chatbots, games                      | Spam filter, recommender systems                  | Face recognition, voice assistants | Sales prediction, analytics, dashboards                      |
| **Role**         | Umbrella concept                             | Learning mechanism                                | Advanced ML method                 | Insights + decision-making                                   |

**6. Key Relationship:**

* **AI** is the overall goal → make systems intelligent.
* **ML** is one way to achieve AI → by learning from data.
* **DL** is a more advanced way within ML → using neural networks.
* **DS** supports all of these → by preparing, analyzing, and interpreting the data needed.

**Question 2: What are the types of machine learning? Describe each with one
real-world example.**

**Answer:**

Machine Learning (ML) is a subset of Artificial Intelligence (AI) that enables machines to learn from data without explicit programming. ML can be broadly classified into three main types based on how the machine learns from data:

**1. Supervised Learning**

* **Definition:**
  In supervised learning, the model is trained on a labeled dataset, which means that each input data has a corresponding correct output. The goal is for the model to learn the mapping between inputs and outputs, so it can predict outcomes for new, unseen data.
* **How it works:**
  The algorithm compares its predicted output with the true output (labels) and adjusts itself to minimize errors.
* **Techniques/Algorithms:**
  Linear Regression, Logistic Regression, Decision Trees, Random Forests, Support Vector Machines (SVM).
* **Real-world Example:**

  * **Email Spam Detection:** The model is trained with labeled emails (spam or not spam) and then predicts whether new incoming emails are spam.

**2. Unsupervised Learning**

* **Definition:**
  In unsupervised learning, the model is trained on unlabeled data, meaning there are no predefined outputs. The goal is to find hidden patterns, structures, or relationships in the data.
* **How it works:**
  The algorithm groups or organizes the data based on similarities or differences without prior knowledge of the outcomes.
* **Techniques/Algorithms:**
  K-Means Clustering, Hierarchical Clustering, Principal Component Analysis (PCA), Association Rules.
* **Real-world Example:**

  * **Customer Segmentation:** Businesses group customers with similar purchasing behavior to design targeted marketing campaigns.

**3. Reinforcement Learning**

* **Definition:**
  In reinforcement learning, the model learns by interacting with the environment. It receives feedback in the form of rewards or penalties for the actions it takes. The goal is to learn a strategy that maximizes cumulative rewards over time.
* **How it works:**
  The agent explores the environment, makes decisions, receives rewards or penalties, and adjusts its actions to improve performance.
* **Techniques/Algorithms:**
  Q-Learning, Deep Q-Networks, Policy Gradient Methods.
* **Real-world Example:**

  * **Self-driving Cars:** The car learns to drive safely by taking actions (steering, braking) and receiving feedback (rewards for staying on the road, penalties for collisions).

**4. Semi-Supervised Learning**

* **Definition:**
  Semi-supervised learning is a combination of supervised and unsupervised learning. It uses a small amount of labeled data with a large amount of unlabeled data to improve learning accuracy.
* **Real-world Example:**

  * **Medical Diagnosis:** Only a few patient records are labeled with disease outcomes, but the algorithm can still learn patterns from the large pool of unlabeled records.

**5. Key Differences (Quick Comparison Table):**

| Type                   | Data Required                   | Goal                               | Example               |
| ---------------------- | ------------------------------- | ---------------------------------- | --------------------- |
| Supervised Learning    | Labeled                         | Predict output for new data        | Spam detection        |
| Unsupervised Learning  | Unlabeled                       | Find hidden patterns/clusters      | Customer segmentation |
| Reinforcement Learning | Feedback-based                  | Learn strategy to maximize rewards | Self-driving cars     |
| Semi-Supervised        | Some labeled + mostly unlabeled | Improve learning with few labels   | Medical diagnosis     |

**6. Summary:**

* **Supervised learning** → “learn with a teacher” (known labels).
* **Unsupervised learning** → “find patterns without a teacher.”
* **Reinforcement learning** → “learn by trial and error.”
* **Semi-supervised learning** → “use a mix of labeled and unlabeled data.”

**Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.**

**Answer:**

**1. Introduction**

In machine learning, the ultimate goal is not just to perform well on training data but to generalize to new, unseen data. Two common problems occur:

* **Overfitting** → Model is too complex.
* **Underfitting** → Model is too simple.

To understand these, we use the concept of the bias-variance tradeoff, which explains how model complexity affects accuracy.

**2. Overfitting**

* **Definition:**
  Overfitting occurs when a model learns the training data too well, including random noise and unnecessary details. It performs well on training data but fails on test data.

* **Characteristics:**

  * Training accuracy: Very high.
  * Test accuracy: Low.

* **Causes:**

  1. Very complex models (deep trees, high-degree polynomials, neural networks without control).
  2. Too many features (irrelevant variables).
  3. Small dataset size compared to model complexity.
  4. Too many training epochs in neural networks.

* **Consequences:**

  * Poor generalization.
  * Wasted computation on noise.

* **Real-world Analogy:**
  A student memorizes answers word-for-word from last year’s exam instead of understanding concepts. If the teacher asks a slightly new question, the student fails.

* **Example:**

  * In medical diagnosis, a model memorizes patient training data instead of learning general symptoms. It fails for new patients.

* **Solutions:**

  * Collect more data.
  * Reduce complexity (pruning trees, using fewer features).
  * Regularization techniques (L1, L2).
  * Early stopping in neural networks.
  * Cross-validation for better testing.

**3. Underfitting**

* **Definition:**
  Underfitting happens when a model is too simple and cannot capture the true structure of the data.

* **Characteristics:**

  * Training accuracy: Low.
  * Test accuracy: Low.

* **Causes:**

  1. Too simple models (linear regression on non-linear data).
  2. Very few features used.
  3. Over-regularization (too much penalty reduces learning).
  4. Insufficient training time.

* **Consequences:**

  * Model ignores important patterns.
  * Predictions are inaccurate even on training data.

* **Real-world Analogy:**
  A student only reads chapter titles without learning details. They cannot answer simple or complex questions.

* **Example:**

  * Predicting house prices using only "square feet" as input, ignoring other factors like location, number of rooms, etc.

* **Solutions:**

  * Use more features.
  * Increase model complexity (polynomial regression, deeper networks).
  * Reduce regularization.
  * Train longer with proper tuning.

**4. Bias and Variance**

* **Bias:**

  * Systematic error due to oversimplified assumptions.
  * High bias → Underfitting.

* **Variance:**

  * Sensitivity of model to small fluctuations in training data.
  * High variance → Overfitting.

**5. Bias-Variance Tradeoff**

* **Concept:**
  There is a natural tension between bias and variance:

  * **Too simple model** → High bias, low variance → Underfitting.
  * **Too complex model** → Low bias, high variance → Overfitting.
  * **Best model** → Balanced bias and variance → Good generalization.

* **Daily Analogy:**

  * High bias: A blunt knife that always cuts poorly (never accurate).
  * High variance: A sharp knife that slips easily and cuts wrongly if slightly misplaced.
  * Tradeoff: A properly balanced knife cuts smoothly and safely.

**6. Combined Example**

Suppose we predict exam marks from study hours:

* **Underfitting:** Model predicts same marks for everyone (ignores pattern).
* **Overfitting:** Model makes a zigzag curve passing through every training point (even noise).
* **Good Tradeoff:** Model captures general trend (more study hours → more marks), fits training and test well.

**7. Importance in Machine Learning**

* Helps in designing models that generalize.
* Prevents waste of resources by avoiding extreme complexity.
* Ensures models are reliable in real-world applications like:

  * Stock prediction.
  * Weather forecasting.
  * Medical diagnosis.
  * Fraud detection.

**8. Conclusion**

* **Overfitting:** Model too complex → High variance → Memorizes data.
* **Underfitting:** Model too simple → High bias → Misses patterns.
* **Bias-Variance Tradeoff:** Balance is required for optimal performance.

A good ML model should not only learn from training data but also perform well on unseen data, ensuring generalization.

**Question 4: What are outliers in a dataset, and list three common techniques for handling them.**

**Answer:**

**1. Introduction - What are Outliers?**

In data analysis, an outlier is a data point that is significantly different from the rest of the observations.

* It lies far away from the central values (mean/median) of a dataset.
* Outliers can occur due to measurement errors, data entry mistakes, or natural variability.

**2. Why Outliers Matter?**

* **Negative impact:**

  * Can distort statistical measures (mean, variance).
  * Can mislead machine learning models, causing poor predictions.
* **Positive impact:**

  * Sometimes represent important discoveries (e.g., fraud detection, rare diseases).

Example: If most students scored between 40-80, but one scored 0 and another 100, these two may be considered outliers.

**3. Causes of Outliers**

1. **Human/Data entry error** → Typing 1000 instead of 100.
2. **Instrument error** → Faulty sensor giving abnormal readings.
3. **Sampling issues** → Data collected from unusual sources.
4. **Natural variation** → Rare but valid cases (e.g., tallest person in a dataset).

**4. Common Techniques for Handling Outliers**

**(i) Removal (Deleting Outliers)**

* If outlier is due to error/noise, simply remove it.
* Useful in large datasets where few missing values don’t affect results.
* **Example:** Removing temperature values like -500°C from a climate dataset.

**(ii) Transformation**

* Apply mathematical functions to reduce the effect of outliers.
* Common transformations:

  * **Logarithmic**
  * **Square root**
  * **Z-score normalization**
* **Example:** Income data often has extreme values; applying log transformation compresses them.

**(iii) Imputation / Capping (Winsorization)**

* Replace extreme values with:

  * Mean/median of the dataset.
  * A percentile value (e.g., 95th percentile).
* **Example:** Replace very high house prices with the 95th percentile value to avoid distortion.

**5. Other Advanced Methods**

* **Clustering methods** → Detect points far from main clusters.
* **Machine learning approaches** → Isolation Forest, DBSCAN.
* **IQR method** → Mark values outside Q1 – 1.5*IQR and Q3 + 1.5*IQR as outliers.

**6. Real-world Examples**

* **Finance:** Sudden spike in credit card spending may indicate fraud.
* **Healthcare:** Abnormally high blood pressure readings may indicate error or rare condition.
* **E-commerce:** One product suddenly getting 10,000 returns may highlight faulty batches.

**7. Conclusion**

Outliers are unusual data points that can either mislead models or provide valuable insights.

* They must be carefully identified and handled using methods such as removal, transformation, or imputation.
* The choice of technique depends on whether the outlier is error or valid rare case.

**Question 5: Explain the process of handling missing values and mention one
imputation technique for numerical and one for categorical data.**

**Answer:**
**1. Introduction - Missing Values**

In datasets, missing values occur when some information is not recorded or is unavailable.

* Example: In a medical dataset, some patients may not report their weight or blood pressure.
* Missing values can cause biased analysis, reduce model accuracy, or lead to incorrect conclusions.

**2. Causes of Missing Data**

1. **Human error** → Data entry mistakes.
2. **System failure** → Sensor or software didn’t record data.
3. **Non-response** → Participants skip questions in surveys.
4. **Data corruption** → Loss of records during transfer/storage.

**3. Why Handle Missing Values?**

* Many algorithms (like SVM, logistic regression) cannot handle missing values directly.
* Missing data reduces dataset quality and affects predictive power.

**4. Process of Handling Missing Values**

**Step 1: Identify Missing Values**

* Use methods like `.isnull()` or `.info()` in Python (Pandas).

**Step 2: Analyze Missingness**

* **MCAR (Missing Completely at Random):** Missing has no pattern.
* **MAR (Missing at Random):** Missing depends on other variables.
* **MNAR (Missing Not at Random):** Missing depends on itself (e.g., people with high income may hide income).

**Step 3: Choose a Strategy**

* **Deletion Methods**

  * *Remove rows* with missing values (useful if few rows are affected).
  * *Remove columns* if too many values are missing.
* **Imputation Methods**

  * Replace missing data with estimated values (mean, median, mode, etc.).

**5. Imputation Techniques**

**(a) Numerical Data Imputation**

* **Mean/Median Imputation**

  * Replace missing numerical values with the column’s mean or median.
  * Example: If "Age" values are missing, replace with average/median age.
  * Median is preferred when data is skewed.

**(b) Categorical Data Imputation**

* **Mode Imputation**

  * Replace missing categorical values with the most frequent category.
  * Example: If "Gender" is missing, replace with the most common value (e.g., "Female").

**6. Advanced Imputation Methods**

* **KNN Imputation** → Finds nearest neighbors and uses them to fill missing values.
* **Regression Imputation** → Predict missing value using other features.
* **Multiple Imputation** → Generates multiple possible values and averages them.

**7. Real-world Examples**

* **Banking:** Missing salary values in loan applications can be filled with median salary of similar profiles.
* **E-commerce:** Missing product categories can be imputed using mode.
* **Healthcare:** Missing patient weight can be replaced with mean weight for the age group.

**8. Conclusion**

Handling missing values is critical for reliable data analysis and machine learning models.

* First identify missingness, then choose an appropriate strategy.
* For numerical data, mean/median imputation is common.
* For categorical data, mode imputation is widely used.
* Advanced methods can be applied when accuracy is more important.

```python
#Question 6: Write a Python program that:
#● Creates a synthetic imbalanced dataset with make_classification() from sklearn.datasets.
#● Prints the class distribution.
#(Include your Python code and output in the code box below.)
#Answer:

# Import required libraries
from sklearn.datasets import make_classification
from collections import Counter
import matplotlib.pyplot as plt

# Function to create and analyze imbalanced dataset
def create_imbalanced_dataset(n_samples=1000, weights=[0.90], random_state=42):
    """
    Creates a synthetic imbalanced dataset and displays class distribution & visualization.

    Parameters:
    -----------
    n_samples : int
        Total number of data points (default = 1000)
    weights : list
        Proportion of each class (default = [0.90] → 90% class 0, 10% class 1)
    random_state : int
        Random seed for reproducibility

    Returns:
    --------
    X, y : dataset features and labels
    """

    # Step 1: Generate dataset
    X, y = make_classification(
        n_samples=n_samples,
        n_features=2,
        n_redundant=0,
        n_clusters_per_class=1,
        weights=weights,
        random_state=random_state
    )

    # Step 2: Print class distribution
    print("Class Distribution in the Synthetic Dataset:")
    class_distribution = Counter(y)
    for label, count in class_distribution.items():
        percentage = (count / len(y)) * 100
        print(f"  Class {label}: {count} samples ({percentage:.2f}%)")

    # Step 3: Visualization
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", alpha=0.6, edgecolors="k")
    plt.title("Synthetic Imbalanced Dataset (Class 0 vs Class 1)", fontsize=14)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()

    return X, y  # Return dataset for further use


# Call the function
X, y = create_imbalanced_dataset()
```

```python
#Question 8: Write a Python script to:
#● Generate 1000 samples from a normal distribution.
#● Introduce 50 random missing values.
#● Fill missing values with the column mean.
#● Plot a histogram before and after imputation.
#(Include your Python code and output in the code box below.)
#Answer:


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def generate_and_impute_normal_data(samples=1000, mean=50, std=10, missing_count=50, random_state=42):
    """
    Function to generate normal distribution data, introduce missing values,
    impute missing values with mean, and plot histograms before and after imputation.

    Parameters:
    - samples: int, number of samples to generate
    - mean: float, mean of the normal distribution
    - std: float, standard deviation of the normal distribution
    - missing_count: int, number of random missing values to introduce
    - random_state: int, seed for reproducibility

    Returns:
    - df: DataFrame with imputed values
    """
    # Set seed
    np.random.seed(random_state)

    # Step 1: Generate normal distribution data
    data = np.random.normal(loc=mean, scale=std, size=samples)
    df = pd.DataFrame(data, columns=["Values"])

    # Step 2: Introduce random missing values
    missing_indices = np.random.choice(df.index, size=missing_count, replace=False)
    df.loc[missing_indices, "Values"] = np.nan

    # Step 3: Plot histogram BEFORE imputation
    plt.figure(figsize=(10,4))
    plt.hist(df["Values"], bins=30, color="skyblue", edgecolor="black")
    plt.title("Histogram Before Imputation")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.show()

    # Step 4: Impute missing values with column mean
    mean_value = df["Values"].mean()
    df["Values"].fillna(mean_value, inplace=True)

    # Step 5: Plot histogram AFTER imputation
    plt.figure(figsize=(10,4))
    plt.hist(df["Values"], bins=30, color="lightgreen", edgecolor="black")
    plt.title("Histogram After Imputation")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.show()

    return df

# --- Call the function ---
imputed_df = generate_and_impute_normal_data()
print(imputed_df.head(10))


```

```python
#Question 10: You are working as a data scientist for a retail company. You receive a customer transaction dataset that contains:
#● Missing ages,
#● Outliers in transaction amount,
#● A highly imbalanced target (fraud vs. non-fraud),
#● Categorical variables like payment method.
#Explain the step-by-step data preparation plan you’d follow before training a machine learning model.
#Include how you’d address missing data, outliers, imbalance, and encoding.
#(Include your Python code and output in the code box below.)
#Answer:

# Data Preparation & Model Training
# Retail Fraud Detection Example

import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils import resample

# -------------------------------
# Step 0: Create a synthetic dataset
# -------------------------------
data = pd.DataFrame({
    "Age": [25, np.nan, 40, 35, np.nan, 50, 28],
    "TransactionAmount": [100, 20000, 150, 120, 30000, 400, 500],
    "PaymentMethod": ["Card", "Cash", "UPI", "Card", "Cash", "Wallet", "UPI"],
    "Fraud": [0, 0, 0, 0, 1, 0, 1]
})

print("Original Dataset:\n", data)

# -------------------------------
# Step 1: Handle Missing Values (Age)
# -------------------------------
# Using median for numerical column 'Age'
imputer = SimpleImputer(strategy="median")
data["Age"] = imputer.fit_transform(data[["Age"]])

# -------------------------------
# Step 2: Handle Outliers (Transaction Amount) using IQR
# -------------------------------
Q1, Q3 = data["TransactionAmount"].quantile([0.25, 0.75])
IQR = Q3 - Q1
lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR

# Cap values outside IQR
data["TransactionAmount"] = np.where(
    data["TransactionAmount"] > upper, upper,
    np.where(data["TransactionAmount"] < lower, lower, data["TransactionAmount"])
)

# -------------------------------
# Step 3: Handle Imbalanced Target (Fraud)
# -------------------------------
majority = data[data.Fraud == 0]
minority = data[data.Fraud == 1]

# Upsample minority class
minority_upsampled = resample(minority, replace=True, n_samples=len(majority), random_state=42)
data_balanced = pd.concat([majority, minority_upsampled])

# Shuffle dataset after resampling
data_balanced = data_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# -------------------------------
# Step 4: Encode Categorical Variable (PaymentMethod)
# -------------------------------
encoder = OneHotEncoder(drop="first", sparse_output=False)
encoded = encoder.fit_transform(data_balanced[["PaymentMethod"]])
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(["PaymentMethod"]))

# Drop original column and add encoded columns
data_balanced = pd.concat([data_balanced.drop("PaymentMethod", axis=1), encoded_df], axis=1)

# -------------------------------
# Step 5: Feature Scaling (Age & TransactionAmount)
# -------------------------------
scaler = StandardScaler()
data_balanced[["Age", "TransactionAmount"]] = scaler.fit_transform(data_balanced[["Age", "TransactionAmount"]])

print("\nProcessed Dataset:\n", data_balanced)

# -------------------------------
# Step 6: Train-Test Split
# -------------------------------
X = data_balanced.drop("Fraud", axis=1)
y = data_balanced["Fraud"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# -------------------------------
# Step 7: Train Random Forest Classifier
# -------------------------------
model = RandomForestClassifier(class_weight="balanced", random_state=42)
model.fit(X_train, y_train)

print("\nModel trained successfully with prepared data!")

# Optional: Check model performance
accuracy = model.score(X_test, y_test)
print("Test Accuracy:", accuracy)

```

First, I handle missing ages using median imputation. Next, I identify and cap outliers in TransactionAmount using IQR. Then, I balance the target variable using upsampling. Categorical variables like PaymentMethod are encoded using OneHotEncoder, and numerical features are scaled using StandardScaler. Finally, the dataset is split into training and test sets, and a RandomForestClassifier is trained on the prepared data.
